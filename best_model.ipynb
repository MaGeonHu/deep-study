```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os


df_pre = pd.read_csv('./dataset/wine.csv',header=None)
df = df_pre.sample(frac=0.15)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
np.random.seed(3)
tf.random.set_seed(3)


model = Sequential()
model.add(Dense(30,input_dim = 12 ,activation = 'relu'))
model.add(Dense(12,activation = 'relu'))
model.add(Dense(8 ,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])


history = model.fit(X,y,validation_split=0.33,epochs=100, batch_size= 50)
```

```python
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']

plt.plot(val_loss,"o", c='red' ,label='val_loss',markersize = 3)
plt.plot(val_acc,"o", c='blue', label = 'val_acc',markersize = 3)
plt.xlabel('epoch')
plt.ylabel('value')
plt.legend()
plt.show()
```

```python
np.random.seed(3)
tf.random.set_seed(3)


model_dir = './model/'
if not os.path.exists(model_dir): 
    os.mkdir(model_dir)

modelpath = './model/{epoch:02d}-{val_loss:.4f}.keras'
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)



model = Sequential()
model.add(Dense(30,input_dim = 12 ,activation = 'relu'))
model.add(Dense(12,activation = 'relu'))
model.add(Dense(8 ,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])


history = model.fit(X,y,validation_split=0.33,epochs=3500, batch_size= 500 ,callbacks=[checkpointer])
```

```python
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']

plt.plot(val_loss,"o", c='red' ,label='val_loss',markersize = 3)
plt.plot(val_acc,"o", c='blue', label = 'val_acc',markersize = 3)
plt.xlabel('epoch')
plt.ylabel('value')
plt.legend()
plt.show()
```

```python
from keras.callbacks import EarlyStopping

np.random.seed(3)
tf.random.set_seed(3)


model_dir = './model/'
if not os.path.exists(model_dir): 
    os.mkdir(model_dir)

modelpath = './model/{epoch:02d}-{val_loss:.4f}.keras'
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience = 50)


model = Sequential()
model.add(Dense(30,input_dim = 12 ,activation = 'relu'))
model.add(Dense(12,activation = 'relu'))
model.add(Dense(8 ,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = ['accuracy'])


history = model.fit(X,y,validation_split=0.33,epochs=3500, batch_size= 500 ,callbacks=[checkpointer,early_stopping])
```

```python
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']

plt.plot(val_loss,"o", c='red' ,label='val_loss',markersize = 3)
plt.plot(val_acc,"o", c='blue', label = 'val_acc',markersize = 3)
plt.xlabel('epoch')
plt.ylabel('value')
plt.legend()
plt.show()
```
